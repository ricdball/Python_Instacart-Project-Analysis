# If-Statements with User-Defined Functions

```python
# Creating subset of first 1000000 rows
df=df_ords_prods_merge[:1000000]
```

```python
# Creating if-statement for new price_label column
def price_label(row):
    if row['prices']<=5:
        return 'Low-range product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-range product'
    elif row['prices']>15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
# applying defined user-defined if-statement to df of first 1000000 rows subset
df['price_range']=df.apply(price_label,axis=1)
```

```python
# Running value_counts to check result values in new column
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

# If-Statements with For-Loops

## Objective: create for-loop that helps summarize how busy each day of the week is

```python
# Getting frequency of orders during the week
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Creating for-loop to have new column summarizing how busy each day of the week is

result=[]
for value in df_ords_prods_merge["orders_day_of_week"]:
    if value==0:
        result.append("Busiest day")
    elif value==4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
# Running for-loop result
result
```

```python
# Creating column busiest day
df_ords_prods_merge['busiest_day']=result
```

```python
# Running value_counts of new busiest day column
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

#  2.Clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this

```python
df_ords_prods_merge.head()
```

```python
# Getting frequency of orders_day_of_week
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

## 0=Saturday/1=Sunday/2=Monday/3=Tuesday/4=Wednesday/5=Thursday/6=Friday
## Two busiest days of the week: Saturday & Sunday
## Two slowest days of the week: Tuesday & Wednesday

## Creating for-loop based on above found two busiest days and two slowest days

```python
# Creating for-loop to show the top two busiest and bottom two slowest days mentioned above

result_2=[]
for value in df_ords_prods_merge["orders_day_of_week"]:
    if value in [0,1]:
        result_2.append("Busiest days")
    elif value in [3,4]:
        result_2.append("Slowest days")
    else:
        result_2.append("Regularly busy")
```

```python
# Running result of new for-loop
result_2
```

```python
## Creating column "busiest_days" to show top two busiest days and bottom two slowest days while still keeping column to show the bottom preforming days
df_ords_prods_merge['busiest_days']=result_2
```

# 4.When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Getting frequency of column orders_hour_of_day
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Creating for-loop to show periods of time when orders are most and fewest
# values are broken into even categories 24hrs in a day for 3 categories of Most orders, Average orders, and Fewest Orders giving 8 in each category

result_3=[]
for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [10,11,14,15,13,12,16,9]:
        result_3.append("Most orders")
    elif value in [17,8,18,19,20,7,21,22]:
        result_3.append("Average orders")
    elif value in [23,6,0,1,5,2,4,3]:
        result_3.append("Fewest orders")
    else:
        result_3.append("No orders")
```

```python
# Running for-loop result_3 list
result_3
```

```python
# Creating new column "busiest_period_of_day"
df_ords_prods_merge['busiest_period_of_day']=result_3
```
