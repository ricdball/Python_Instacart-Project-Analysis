# 6.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
## Criteria 1: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
## Criteria 2: If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating new column for average of customer spending
ords_prods_merge['spending_average']=ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

## Creating flag column to identify type of spending customer.

```python
ords_prods_merge.loc[ords_prods_merge['spending_average']<10,'type_of_spender']='Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spending_average']>=10,'type_of_spender']='High Spender'
```

```python
# Checking value counts of newly created column "type_of_spender flag
ords_prods_merge['type_of_spender'].value_counts(dropna=False)
```

## Results show that Instacart customers are mainly low spending customers.

# 7.In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating new column for user/customer frequency median
ords_prods_merge['user_frequency']=ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

## creating frequency flag of customers

```python
ords_prods_merge.loc[ords_prods_merge['user_frequency']<=10,'customer_frequency_type']='Frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['user_frequency']>10)&(ords_prods_merge['user_frequency']<=20),'customer_frequency_type']='Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_frequency']>20,'customer_frequency_type']='Non-frequent Customer'
```

```python
# Checking value counts of new customer freqency column
ords_prods_merge['customer_frequency_type'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id','user_frequency','customer_frequency_type']].head(50)
```
